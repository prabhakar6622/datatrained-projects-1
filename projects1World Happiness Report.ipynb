```python
def show_nulls(df):
    total_nan = df.isnull().sum()
    perc_nan = total_nan/len(df)*100
    
    return total_nan, perc_nan
```

```python
print('Total nul', show_nulls(df=df)[0])
```

```python
print('Percentage of null', show_nulls(df=df)[1])
```
